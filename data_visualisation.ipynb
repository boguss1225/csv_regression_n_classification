```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("heart_attack_dataset/heart.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
category_var=["sex",'cp',"fbs","exng","restecg","thall","caa","slp",'output']
continuous_var=["age","trtbps","chol","thalachh","oldpeak"]
```

```python
plt.figure(figsize=(18,10))
plt.style.use("ggplot")
sns.countplot(x=df["age"])    #using countplot
plt.title("COUNT OF PATIENTS AGE",fontsize=20)
plt.xlabel("AGE",fontsize=20)
plt.ylabel("COUNT",fontsize=20)
# plt.savefig("age_dist")
```

```python
s=df["sex"].value_counts().reset_index()
df.groupby('sex').size().plot(kind='pie',labels=["Men","Women"], autopct='%.2f')
# plt.savefig("gender_pie")
```

```python
sns.boxplot(data = df[continuous_var])
# plt.savefig("boxplot")
```

```python
plt.figure(figsize=(30,30))

plt.subplot(2,2,1)
sns.histplot(data=df, x='age').set(title="age distribution")

plt.subplot(2,2,2)
sns.histplot(data=df, x='trtbps', color='orange').set(title="blood pressure distribution")

plt.subplot(2,2,3)
sns.histplot(data=df, x='thalachh', color='green').set(title="max heart rate distribution")

plt.subplot(2,2,4)
sns.histplot(data=df, x='oldpeak', color='yellow').set(title="oldpeak distribution")
```

```python
plt.figure(figsize=(18,8))
sns.heatmap(df.corr(),annot=True,cmap="PuBuGn")  #using heatmap here
# plt.savefig("corr")
```

```python
sns.pairplot(data = df[continuous_var+["output"]], hue='output')
# plt.savefig("pair_plot")
```
